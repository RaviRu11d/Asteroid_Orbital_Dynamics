```python
import os
import pathlib
import pandas as pd
```

```python
cols_prop = [
    "Asteroid_id",
    "H",
    "a_p",
    "e_p",
    "sin_i_p",
    "n",
    "g",
    "s",
    "LCE",
    "My"
]
```

```python
core_path = os.getcwd()
prop_df = pd.read_csv("data/lvl1/proper_elements.syn", delimiter='\s+', skiprows=2, names=cols_prop)
```

```python
prop_df.head()
```

```python
cols_fam = [
    "Asteroid_id",
    "H",
    "status",
    "family_1",
    "dv_fam_1",
    "near_1",
    "family_2",
    "dv_fam_2",
    "near_2",
    "rescod"
]
```

```python
family_df = pd.read_csv("data/lvl1/family_classification.famrec", delimiter='\s+', names=cols_fam, comment='%')
```

```python
family_df.head()
```

```python
cols_rms = [
    "Asteroid_id",
    "a_p",
    "e_p",
    "sin_i_p",
    "n",
    "g",
    "s",
    "lam_fit_deg",
    "My"
]
```

```python
RMS_mb_df = pd.read_csv("data/lvl1/RMS_main_belt.sig", delimiter='\s+', names=cols_rms, comment='%')
```

```python
RMS_mb_df['a_p'] = RMS_mb_df['a_p'].map(lambda x: f"{x:.8f}")
RMS_mb_df['e_p'] = RMS_mb_df['e_p'].map(lambda x: f"{x:.7f}")
RMS_mb_df['sin_i_p'] = RMS_mb_df['sin_i_p'].map(lambda x: f"{x:.7f}")
```

```python
RMS_mb_df.head()
```

```python
pathlib.Path(os.path.join(core_path, "data/lvl2")).mkdir(parents=True, exist_ok=True)

prop_df.to_pickle(os.path.join(core_path, "data/lvl2", "prop_df.pkl"), protocol=4)
family_df.to_pickle(os.path.join(core_path, "data/lvl2", "family_df.pkl"), protocol=4)
RMS_mb_df.to_pickle(os.path.join(core_path, "data/lvl2", "RMS_mb_df.pkl"), protocol=4)
```
