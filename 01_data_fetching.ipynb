```python
import gzip
import os
import pathlib
import shutil
import urllib.request
```

```python
core_path = os.getcwd()
pathlib.Path(os.path.join(core_path, "data/lvl1")).mkdir(parents=True, exist_ok=True)
```

```python
files_to_dl = {
    'proper_elements': {
        'url': 'https://newton.spacedys.com/~astdys2/propsynth/all.syn'
    },

    'family_classification': {
        'url': 'https://newton.spacedys.com/~astdys2/propsynth/all_tro.famrec'
    },

    'RMS_main_belt': {
        'url': 'https://newton.spacedys.com/~astdys2/propsynth/all.sig'
    }
}
```

```python
for dl_key in files_to_dl:

    split = urllib.parse.urlsplit(
        files_to_dl[dl_key]["url"]
    )

    filename = pathlib.Path(
        os.path.join(
            core_path,
            "data/lvl1/",
            split.path.split("/")[-1]
        )
    )

    # Download if missing
    if not filename.is_file():

        print(f"Downloading: {files_to_dl[dl_key]['url']}")

        downl_file_path, _ = urllib.request.urlretrieve(
            url=files_to_dl[dl_key]['url'],
            filename=filename
        )

        print(f"Saved → {downl_file_path}")

    else:

        print(f"Already exists → {filename}")
```
